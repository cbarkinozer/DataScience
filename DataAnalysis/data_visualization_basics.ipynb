```python
#For array operations
import numpy as np
#A fundamental plotting library
import matplotlib.pyplot as plt
```

```python
#Shows the visualization inline
%matplotlib inline 
```

```python
#Divide 0 to 10 to 100 (this is our fake data)
x= np.linspace(0,10,100)
print(x)
```

```python
#Plotting
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
#Axes labels
plt.xlabel('x axes')
plt.ylabel('y axes')
#Legend to understand what colors represent
plt.legend(('sin','cos'))
```

```python
#Plot figure
fig=plt.figure()
#We can make subplot(row,column,index)
#If you do not write subplot it will be overplotted
plt.subplot(2,1,1)
plt.plot(x,np.sin(x),'-')
plt.xlabel('x axes')
plt.ylabel('y axes')
plt.legend(("sin","sin"))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x),'--',linewidth=4,color='red')
plt.xlabel('x axes')
plt.ylabel('y axes')
plt.legend(("cos","cos"))
```

```python
#Saves figure as a file
fig.savefig('data.png')
```

```python
import seaborn as sns
import pandas as pd
```

```python
#Download popular datasets from web
iris= sns.load_dataset('iris')
#Show top values
iris.head()
```

```python
#Plot iris dataset's petal_length column
iris['petal_length'].plot()
```

```python
#Diferent specie's by their petal_length
sns.lineplot(data=iris,y='petal_length',x=iris.index,hue='species')
```

```python
#Scatter plotting shows correlation between numerical datasets
#Overplotting is trying to scatter plot too many data which cause a visually meaningless pile visual
#Solutions of overplotting: shrink the size of plots or reducing the transparency of dots
#Canvas settings
fig,ax= plt.subplots(figsize=(10,5))
#Plotting
ax.scatter(iris['sepal_length'],iris['sepal_width'])
#Title of the figure
ax.set_title('Iris Dataset')
#Labels of axes
ax.set_xlabel('sepal_length')
ax.set_xlabel('sepal_width')
```

```python
#Barplot: Shows sum,average,median etc.on a grouped categoric column.
#Do not confuse barplot with histogram.
#If category names are too long, try to plot barplot sideways.
#Prefer dual column barplots over bidirectional charts
```

```python
data=iris.groupby(by='species').mean()['sepal_length']
ax=sns.barplot(x=data.index,y=data.values)
ax.set_title('Sepal Length Means')
ax.set_ylabel('Mean')
```

```python
#Histogram grouped continues values where barplot is for categoric values.
#histograms give an estimate of where the values are concentrated,
#what are the extremes, and whether there are any gaps or unusual values in the dataset
#Bin width= 3.5 * standard derivation/(tertiary root(number of values in data source))
```

```python
sns.histplot(data=iris,x='sepal_width',bins=10)
```

```python
sns.histplot(data=iris,x='sepal_width',bins=20)
```

```python
#Pie charts: shows the percentage on a categorical column
#Do not compare different pie charts, total percentage must be equal to 100
plt.pie(data)
```

```python
#Heatmap: Using colors for showing matrix's value. It is used for finding correlation.
#Correlation is two variables change rate by other each other.
iris.corr()
```

```python
#Annot shows the values
sns.heatmap(iris.corr(),annot=True)
```

```python
#Boxplot: shows the data distribution between groups
#It consists of 5 parts, minimum first quartile, median 3rd quartile and maximum
#It shows variation range and outliers.
sns.boxplot(data=iris,y='sepal_length',x='species')
```
