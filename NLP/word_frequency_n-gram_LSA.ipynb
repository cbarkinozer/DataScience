```python
f= open('tur_mixed_2014_10K-sentences.txt','r',encoding='utf8')
text=f.read()
f.close()
```

Turkish Corpora link(TR,mixed,10k,sentences): https://wortschatz.uni-leipzig.de/en/download/Turkish

```python
#Failed to capture some special characters. I added replace code for them manually. 
import string
clean_text = text.translate(str.maketrans('','', string.punctuation)).replace("’","").replace("”","").replace("»","").replace("“","").lower()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

my_stopwords=set(stopwords.words('turkish'))
```

```python
tokens=nltk.word_tokenize(clean_text)
```

```python
#Non-stop words are added by append
words=[]
for token in tokens:
    if token not in my_stopwords:
        words.append(token)
words=" ".join(words) # Converts list to string to make it tokenizable        
tokens=nltk.word_tokenize(words) # Tokenized because it is a string
```

```python
#List of all words with their frequencies
from collections import Counter
def token_counter(tokens):
    counts=Counter(tokens)
    return counts
vocabulary=token_counter(tokens)
print(vocabulary)
```

```python
#Bigrams
import pandas as pd
count_bigrams = (pd.Series(nltk.ngrams(tokens, 2)).value_counts())
print("Number of bigrams (more than 5):",len(count_bigrams[count_bigrams > 5]))
```

```python
#Trigrams
count_trigrams = (pd.Series(nltk.ngrams(tokens, 3)).value_counts())
print("Number of trigrams (more than 5):",len(count_trigrams[count_trigrams > 5]))
```

```python
bigram_series = (count_bigrams)[:10]
trigram_series = (count_trigrams)[:10]
print(bigram_series)
print(trigram_series)
```

```python
#Blocks truncating of the result
pd.options.display.max_rows = 550
#Auto-detect terminal size
pd.options.display.width = 0
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import Normalizer


vectorizer=CountVectorizer(min_df=1)
dtm=vectorizer.fit_transform(f2)
words=vectorizer.get_feature_names()

words=vectorizer.fit_transform(words)
words=words.asfptype()
lsa=TruncatedSVD(2, algorithm='arpack')

dtm_lsa=lsa.fit_transform(words)
dtm_lsa=Normalizer(copy=False).fit_transform(dtm_lsa)
similarity=np.asarray(np.asmatrix(dtm_lsa)*np.asmatrix(dtm_lsa).T)
x=pd.DataFrame(similarity,index=vectorizer.get_feature_names(),columns=vectorizer.get_feature_names())
print(x)
```
