```python
import nltk
```

```python
f= open('tur_mixed_2014_10K-sentences.txt','r',encoding='utf8')
```

Turkish Corpora link(TR,mixed,10k,sentences): https://wortschatz.uni-leipzig.de/en/download/Turkish

```python
text=f.read()
f.close()
```

```python
import string
clean_text = text.translate(str.maketrans('','', string.punctuation)).lower()
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

my_stopwords=set(stopwords.words('turkish'))
```

```python
tokens=nltk.word_tokenize(clean_text)
```

```python
words=[]
for token in tokens:
    if token not in my_stopwords:
        words.append(token)
```

```python
nltk.download('averaged_perceptron_tagger')
from collections import Counter
def nltk_postag(tokens):
    tags=nltk.pos_tag(tokens)
    counts=Counter(tags)
    return counts
pos=nltk_postag(tokens)
print(pos)
```

```python
import zeyrek
pos=zeyrek.MorphAnalyzer()
zeyrek_tags = []
for token in tokens:
    zeyrek_tags.append(pos.analyze(token))
print(zeyrek_tags)
#Surpressing warnings
import warnings
warnings.filterwarnings("ignore")
#Keeps working because there are 10k sentence
#Keyboard interrupt: I+I
```

Zeyrek lemmatize the tokens before postagging therefore there are several same words with different tags.<br />
"genç" is NN in the first pos-tagger. For the second tagger "genç" is adj and noun, and "Genç" word also exist and it is a noun.<br />
"oymaklarla" is noun in the first pos-tagger. For the second tagger "oymaklarla" is lemmatized and become "oymak" therefore it is tagged as a verb.<br />
"diğer" is nn in the first tagger. For the second tagger "diğer" is adj.<br />
